```python
# Import library
import pandas as pd
import numpy as np
```

```python
#set up display
pd.set_option('display.max_rows', None)
```

```python
# read csv Data Gold SJC
df_Gold = pd.read_csv('Data Gold SJC.csv')
```

```python
df_Gold.shape
```

```python
df_Gold.head(10)
```

```python
df_Gold.tail(10)
```

```python
# replace '-' => '/' in DATE
df_Gold['DATE'] = df_Gold['DATE'].str.replace('-', '/')
```

```python
df_Gold.head(5)
```

```python
df_Gold = df_Gold.drop('Unnamed: 0', axis=1)
```

```python
# read file Blog Price Gold 
df_Blog = pd.read_csv('Blog Price Gold.csv')
```

```python
df_Blog.head(10)
```

```python
# reversed data frame
df_reversed = df_Blog.iloc[::-1].reset_index(drop=True)
```

```python
# extract text
df_reversed['giavangngay_text'] = df_reversed['TITLE'].str.extract('(Giá vàng ngày)')
df_reversed['giavanghomnay_text'] = df_reversed['TITLE'].str.extract('(Giá vàng hôm nay)')
df_reversed['phiengiaodich_text'] = df_reversed['TITLE'].str.extract('(Phiên giao dịch)')
df_reversed['giavang **'] = df_reversed['TITLE'].str.extract('(Giá vàng \d+)')
```

```python
df_reversed = df_reversed.drop('Unnamed: 0', axis=1)
```

```python
df_reversed.head(5)
```

```python
df_reversed['giavang **'].value_counts()
```

```python
df_Blog = df_reversed[(df_reversed['giavangngay_text'] == 'Giá vàng ngày') | 
                      (df_reversed['giavanghomnay_text'] == 'Giá vàng hôm nay') | 
                      (df_reversed['phiengiaodich_text'] == 'Phiên giao dịch') | 
                      (df_reversed['giavang **'] == 'Giá vàng 20') |
                      (df_reversed['giavang **'] == 'Giá vàng 23')]
```

```python
df_Blog = df_Blog.drop('giavangngay_text', axis=1)
df_Blog = df_Blog.drop('giavanghomnay_text', axis=1)
df_Blog = df_Blog.drop('phiengiaodich_text', axis=1)
df_Blog = df_Blog.drop('giavang **', axis=1
                       
```

```python
# Extract DATE
df_Blog[['DATE']] = df_Blog['TITLE'].str.extract('(\d{1,2}/\d{1,2}(?:/\d{4})?)')
```

```python
df_Blog.head(5)
```

```python
df_Blog.loc[73:73, 'DATE'] = df_Blog.loc[73:73, 'DATE'] + '/2020'
df_Blog.loc[149:173, 'DATE'] = df_Blog.loc[149:173, 'DATE'] + '/2020'
df_Blog.loc[381:582, 'DATE'] = df_Blog.loc[381:582, 'DATE'] + '/2022'
df_Blog.loc[584:1083, 'DATE'] = df_Blog.loc[584:1083, 'DATE'] + '/2023'
df_Blog.loc[1086:1259, 'DATE'] = df_Blog.loc[1086:1259, 'DATE'] + '/2024'
```

```python
df_Gold.shape, df_Blog.shape
```

```python
df_Blog['DATE'] = pd.to_datetime(df_Blog['DATE'], format='%d/%m/%Y').dt.strftime('%d/%m/%Y')
```

```python
# Join df_Gold and df_Blog
result = pd.merge(df_Gold, df_Blog, how='left', on='DATE')
```

```python
result
```

```python
# To csv file
result.to_csv('data.csv')
```
