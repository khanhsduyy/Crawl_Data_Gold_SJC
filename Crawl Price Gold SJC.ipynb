```python
# Check leap year
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False
    
```

```python
# day: 1 -> 01; month: 1-> 01
def format_str(x):
    convert_str = str(x)
    len_str = len(convert_str)
    if(len_str == 1):
        return "0" + convert_str
    return convert_str
```

```python
# Function create date
def Create_StrDate(day, month, year):
    date = str(day) + '-' + str(month) + '-' + str(year)
    
    return date
```

```python
# Function get price gold
def Get_Info():
    td1_text = None
    td2_text = None
    td3_text = None
    td4_text = None
    try:
        tr1_field = driver.find_element(By.XPATH, '//*[@id="main"]/div[1]/div/table/tbody/tr[1]')
        td_field = tr1_field.find_elements(By.CSS_SELECTOR, 'td')
        
        td1_text = td_field[0].text.strip()
        td2_text = td_field[1].text.strip()
        td3_text = td_field[2].text.strip()
        td4_text = td_field[3].text.strip()
    except:
        sleep(0.2)
    return td1_text, td2_text, td3_text, td4_text
```

```python
# Function create list Date
def Create_ListDate(year, day_now, month_now, year_now):
    list_date = []
    
    for i in range(12):
        month = i + 1
        if month == 1 or month == 3 or month == 5 or month == 7 or month == 8 or month == 10 or month == 12:
            for j in range(31):
                day = j + 1
                
                str_date = Create_StrDate(format_str(day),
                                          format_str(month),
                                          format_str(year))
                list_date.append(str_date)
                if day == day_now and month == month_now:
                    break
            if month == month_now and year == year_now:
                break
        elif month == 2:
            if is_leap_year(year):
                for m in range(29):
                    day = m + 1
                    
                    str_date = Create_StrDate(format_str(day), format_str(month), format_str(year))
                    list_date.append(str_date)
            else:
                for n in range(28):
                    day = n + 1
                    
                    str_date = Create_StrDate(format_str(day), format_str(month), format_str(year))
                    list_date.append(str_date)
        else:
            for j in range(30):
                day = j + 1
                
                str_date = Create_StrDate(format_str(day), format_str(month), format_str(year))
                list_date.append(str_date)
    return list_date
```

```python
# Crear list date 2020 - Now
list_DateFull = []
for year in range(2020, 2025):
    list_DateYear = Create_ListDate(year, day_now, month_now, year_now)
    
    list_DateFull += list_DateYear
    
```

```python
# Create link to Price Gold SJC

ori_link = 'https://webgia.com/gia-vang/sjc/'
tail_link = '.html'

link_full = []
def Create_Link(list_date):
    for date in list_date:
        link = ori_link + date + tail_link
        
        link_full.append(link)
    return link_full
```

```python
url_Price_Gold = Create_Link(list_DateFull)
url = pd.DataFrame(url_Price_Gold)
url.to_csv('url_PriceGold.csv')

print('Create file csv success')
```

```python
# Get Bid_price and price
data = []
def Get_Data(list_url):
    SE_NUMBER = None
    TIME = None
    BID_PRICE = None
    PRICE = None
    for url in list_url:
        driver.get(url)
        
        SE_NUMBER, TIME, BID_PRICE, PRICE = Get_Info()
        
        data.append({'SE_NUMBER': SE_NUMBER, 'TIME': TIME, 'BID_PRICE': BID_PRICE, 'PRICE': PRICE})
        
    return data
        
```
